```python
def dist(t, p):
    sum = 0
    if np.size(p) == 1:
        sum += (t-p)**2
    else:
        for i in range(np.size(p)):
            sum += (t[i]-p[i])**2
    return np.sqrt(sum)

def nearest_mean_classifier(X_train, y_train, X_test, y_test, num_categs, metric, plot=False):
    ##### TRAINING PHASE #####
    categ_means_train = []
    categ_dists_train = []
    y_hat_train = []

    for i in range(num_categs):
        categ_means_train.append(np.average(X_train[y_train == i+1], axis=0))

    if plot:
        pDB(X_train, y_train, np.array(categ_means_train))
    
    for x in X_train:
        dists = []
        for i in range(num_categs):
            dists.append(metric(categ_means_train[i], x))

        categ_dists_train.append(dists)
        min_dist = min(dists)
        y_hat_train.append(dists.index(min_dist)+1)

    num = 0
    denom = 0
    for (y, yh) in zip(y_train, y_hat_train):
        denom += 1
        if y == yh:
            num += 1
    
    train_error_rate = 1-(num/denom)

    ##### TESTING PHASE #####
    categ_dists_test = []
    y_hat_test = []

    for x in X_test:
        dists = []
        for i in range(num_categs):
            dists.append(metric(categ_means_train[i], x))

        categ_dists_test.append(dists)
        min_dist = min(dists)
        y_hat_test.append(dists.index(min_dist)+1)

    num = 0
    denom = 0
    for (y, yh) in zip(y_test, y_hat_test):
        denom += 1
        if y == yh:
            num += 1
    
    test_error_rate = 1 - (num/denom)
    return train_error_rate, test_error_rate
```

```python
X_train_syn1 = syn1_train[:, :2]
y_train_syn1 = syn1_train[:, 2]

X_test_syn1 = syn1_test[:, :2]
y_test_syn1 = syn1_test[:, 2]

train_error_rate1, test_error_rate1 = nearest_mean_classifier(X_train_syn1, y_train_syn1, X_test_syn1, y_test_syn1, 2, dist, True)
print("The training error rate on the first synthetic dataset is: " + str(train_error_rate1))
print("The testing error rate on the first synthetic dataset is: " + str(test_error_rate1))
```

```python
wine_train_df = pd.read_csv('wine_train.csv', header=None)
wine_test_df = pd.read_csv('wine_test.csv', header=None)
wine_train = wine_train_df.to_numpy()
wine_test = wine_test_df.to_numpy()

X_train_wine = wine_train[:, :2]
y_train_wine = wine_train[:, 13]

X_test_wine = wine_test[:, :2]
y_test_wine = wine_test[:, 13]

wine_train_error_rate, wine_test_error_rate = nearest_mean_classifier(X_train_wine, y_train_wine, X_test_wine, y_test_wine, 3, dist, True)
print("The training error rate on the wine dataset is: " + str(wine_train_error_rate))
print("The testing error rate on the wine dataset is: " + str(wine_test_error_rate))
```

## Part d)

```python
wine_train_error_rates = []

for i in range(np.size(wine_train, axis=1)-1):
    wine_train_error_rate = nearest_mean_classifier(wine_train[:, i], y_train_wine, wine_test[:, i], y_test_wine, 3, dist)[0]
    wine_train_error_rates.append(wine_train_error_rate)

train = wine_train_error_rates
min_trains = {}

for i in range(5):
    min_train_val = min(train)
    min_train_index = train.index(min_train_val)

    min_trains[min_train_index] = min_train_val

    del train[min_train_index]

print(min_trains)
```

```python
error_rates = {}

for feat1, feat2 in list(product(min_trains, min_trains)):
    filter = [feat1, feat2]

    wine_train_error_rate = nearest_mean_classifier(wine_train[:, filter], y_train_wine, wine_test[:, filter], y_test_wine, 3, dist)[0]
    error_rates['Train ' + str(feat1) + '+' + str(feat2)] = wine_train_error_rate

print(error_rates)
```

```python
tr = {}
te = {}

for x in list(combinations(range(13), 2)):
    filter = [x[0], x[1]]
    wine_train_error_rate, wine_test_error_rate = nearest_mean_classifier(wine_train[:, filter], y_train_wine, wine_test[:, filter], y_test_wine, 3, dist)
    tr[x] = wine_train_error_rate
    te[x] = wine_test_error_rate

print('The standard deviation of error rates over all possible pairs of training features is: ' + str(np.std(list(tr.values()))))
print('The standard deviation of error rates over all possible pairs of testing features is: ' + str(np.std(list(te.values()))))
print('The minimum training error rate is: ' + str(min(list(tr.values()))) + '.\nThe maximum training error rate is: ' + str(max(list(tr.values()))))
print('The minimum testing error rate is: ' + str(min(list(te.values()))) + '.\nThe maximum testing error rate is: ' + str(max(list(te.values()))))
```
